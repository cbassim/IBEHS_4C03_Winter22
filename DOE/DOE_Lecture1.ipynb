```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
X1=pd.Series([-1,1,-1,1],name='Temp')
X2=pd.Series([-1,-1,1,1], name='Substrate')
X3=pd.Series([1,-1,-1,1])
Y=pd.Series([69,60,64,53],name='Output')
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

DF=pd.concat([X1,X2,X3,Y],axis='columns')
print(DF)
```

```python
fig, ax =plt.subplots(1,2,sharex=True)
sns.lineplot(x="Temp",y="Output",data=DF,ax=ax[0],ci=None)
sns.lineplot(x="Substrate",y="Output",data=DF, ax=ax[1],ci=None)
```

```python
# factorplot, catplot
fig, ax =plt.subplots(1,2,sharex=True)
sns.lineplot(x="Temp",y="Output",hue="Substrate",data=DF,ax=ax[0])
sns.lineplot(x="Substrate",y="Output",hue="Temp",data=DF, ax=ax[1])
```

```python
model1=ols('Y~X1+X2+X3',data=DF)
results1=model1.fit()
print(results1.summary())
```

```python
model2=ols('Y~X1+X2+X1*X2',data=DF)
results2=model2.fit()
print(results2.summary())
```

# Interaction effects

```python
Xa=pd.Series([-1,1,-1,1],name='Temp')
Xb=pd.Series([-1,-1,1,1], name='Substrate')
Xc=pd.Series([1,-1,-1,1])
Yabc=pd.Series([77,79,81,89],name='Output')
```

```python
DFabc=pd.concat([Xa,Xb,Xc,Yabc],axis='columns')
print(DFabc)
```

```python
model3=ols('Yabc~Xa+Xb+Xa*Xb',data=DFabc)
results3=model3.fit()
print(results3.summary())
```

```python
# factorplot, catplot
fig, ax =plt.subplots(1,2,sharex=True)
sns.lineplot(x="Temp",y="Output",hue="Substrate",data=DFabc,ax=ax[0])
sns.lineplot(x="Substrate",y="Output",hue="Temp",data=DFabc, ax=ax[1])
```
